```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
```

```python
req = requests.get('https://www.worldometers.info/coronavirus/country/bangladesh/')
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
elem = soup.select('#maincounter-wrap > div > span')
totalCases = int(elem[0].text.replace(',', ''))
totalDeaths = int(elem[1].text.replace(',', ''))
totalRecovered = int(elem[2].text.replace(',', ''))

data = [totalCases, totalDeaths, totalRecovered]
dataLabels = ['Cases', 'Deaths', 'Recovered']
colors = ['c', 'm', 'y']

plt.pie(data, labels=dataLabels, colors=colors, shadow=True, radius=2, autopct='%1.2f%%', startangle=60)
plt.legend()
plt.show()
```
